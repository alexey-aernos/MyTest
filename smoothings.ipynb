```python
# import holy panda
import pandas as pd
```

```python
# make plots pretty
import matplotlib.pyplot as plt
%matplotlib inline
# list all parameters by "print(plt.rcParams)" command
plt.rcParams['figure.figsize'] = (7.5, 7.5)
plt.rcParams['font.size'] = 15
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.facecolor'] = 'w'
```

```python
# generating range of points
a, b = (1.,2.)
N = 100
n = 1
x = pd.np.linspace(a, b, N+1)
```

```python
# generating step function
y = pd.np.array([0.]*len(x))
y[N/2-n:N/2+n+1]=1
y +=1 
df = pd.DataFrame(zip(x,y), columns=['x','y'])
```

```python
# experimenting with various rolling and ewm smoothings
plt.plot(df.x,df.y, 'k')
plt.plot(df.x,df.y.rolling(5, center=False).mean(),'b', linestyle=' ', marker='o')
plt.plot(df.x,df.y.ewm(5).mean(),'r', linestyle=' ', marker='o')
plt.plot(df.x,df.y.ewm(20).mean(),'g', linestyle=' ', marker='o')
plt.xlim(1.4,1.8)
plt.ylim(0.85,1.65)
plt.title("understanting various smoothings")
plt.xlabel("argument value")
plt.ylabel("function value")
```
