```python
# import pandas for data munching
import pandas as pd
```

```python
# make plots pretty
import matplotlib.pyplot as plt
%matplotlib inline
# list all parameters by "print(plt.rcParams)" command
plt.rcParams['figure.figsize'] = (7.5, 7.5)
plt.rcParams['font.size'] = 15
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.facecolor'] = 'w'
```

```python
#create weekday@10AM index
ts_index = pd.date_range('8/1/2014 10', '8/31/2014', freq='B')
#introduce jitter with 1 hour standard deviation
N = ts_index.size
ts_index += pd.to_timedelta(pd.np.random.normal(size=N), unit='h')
```

```python
# generate abnormality sunday morning with 1 sec interval
ts_index = ts_index.append(pd.date_range('8/24/2014 3:33:33', 
                    periods=120, freq='S')).sort_values()
```

```python
# ths is how transaction look on TS plot
pd.Series(1,ts_index).plot(style='p')
```

```python
# now synthetize data frame with some labels
df = pd.DataFrame(zip(['varganov']*N, ['bl']*N, ts_index))
```

```python
# save synthetic data to disk
df.to_csv('bla.tsv', date_format='%Y%m%d%H%M%S', header=False, sep='\t')
```
